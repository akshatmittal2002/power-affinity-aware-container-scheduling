```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

# Fitting Cubic Power Model

```python
def original_func(x):
    return 0.2 + 0.8 * x**3

x_values = np.linspace(0, 1, 100)
x_values_1 = x_values[x_values <= 0.5]
x_values_2 = x_values[x_values > 0.5]

y_original = original_func(x_values)

model_1 = LinearRegression()
model_2 = LinearRegression()

x_values_1 = x_values_1.reshape(-1, 1)
x_values_2 = x_values_2.reshape(-1, 1)

model_1.fit(x_values_1, original_func(x_values_1))
model_2.fit(x_values_2, original_func(x_values_2))

y_linear_1 = model_1.predict(x_values_1)
y_linear_2 = model_2.predict(x_values_2)

plt.figure(figsize=(7, 3))

plt.plot(x_values, y_original, label='Cubic Power Model', color='red', linewidth = 2)
plt.plot(x_values_1, y_linear_1, label='Linear Approximation', linestyle='--', color='blue', linewidth = 2)
plt.plot(x_values_2, y_linear_2, linestyle='--', color='blue', linewidth = 2)

plt.vlines(0.5, 0, original_func(0.5), colors='black', linestyles='dotted')

plt.xticks([0, 0.5, 1], [r"$u_{min}$", r"$\lambda$", r"$u_{max}$"])
plt.yticks([0.2, 1], [r"$P_{idle}$", r"$P_{max}$"])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.xlabel('Utilization', fontsize=12)
plt.ylabel('Power', fontsize=12)
plt.legend(fontsize=12,edgecolor='k',ncol=2)

# plt.show()
plt.savefig('cubic_linear_fit.pdf',bbox_inches='tight')
```
